### 메타정의서 eng & kor 컬럼명 가지고 오기

```python
file_path = './files_org/표본자료 변수 테이블(2012-2013).xlsx'
df_20TABLE_col= pd.read_excel(file_path,
              sheet_name='20table_명세서일반내역',             #sheet 이름 또는 인덱스 번호 넣으면 해당 시트항목이 나옴
              usecols = 'A,D')
```

```python
file_path = './files_org/표본자료 변수 테이블(2012-2013).xlsx'
df_30TABLE_col= pd.read_excel(file_path,
              sheet_name='30table_진료내역',             #sheet 이름 또는 인덱스 번호 넣으면 해당 시트항목이 나옴
              usecols = 'A,D')
```

```python
file_path = './files_org/표본자료 변수 테이블(2012-2013).xlsx'
df_40TABLE_col= pd.read_excel(file_path,
              sheet_name='40table_상병',             #sheet 이름 또는 인덱스 번호 넣으면 해당 시트항목이 나옴
              usecols = 'A,D')
```

```python
file_path = './files_org/표본자료 변수 테이블(2012-2013).xlsx'
df_53TABLE_col= pd.read_excel(file_path,
              sheet_name='53table_처방전상세내역',             #sheet 이름 또는 인덱스 번호 넣으면 해당 시트항목이 나옴
              usecols = 'A,D')
```

```python
'''
데이터프레임의 칼럼명 = 딕셔너리의 key
    딕셔너리의 value 값으로 대체
'''
```

```python
df_20TABLE_col_list = df_20TABLE_col.to_dict('records')
df_20TABLE_col_list
```

```python
df_20TABLE_col_list = df_20TABLE_col.to_dict('records')
new_dict_20 = {}

for df_1 in df_20TABLE_col_list:

    new_dict_20[df_1['변수명']]=df_1['표준한글컬럼명'] 
          
new_dict_20
    

df_20TABLE_kor = df_20TABLE.rename(columns = new_dict_20)
df_20TABLE_kor
```

```python
df_30TABLE_col_list = df_30TABLE_col.to_dict('records')
new_dict_30 = {}

for df_1 in df_30TABLE_col_list:

    
    new_dict_30[df_1['변수명']] = df_1['표준한글컬럼명']        
new_dict_30
    

df_30TABLE_kor = df_30TABLE.rename(columns = new_dict_30)
df_30TABLE_kor
```

```python
df_40TABLE_col_list = df_40TABLE_col.to_dict('records')
new_dict_40 = {}

for df_1 in df_30TABLE_col_list:

    
    new_dict_40[df_1['변수명']] = df_1['표준한글컬럼명']        
new_dict_40
    

df_40TABLE_kor = df_40TABLE.rename(columns = new_dict_40)
df_40TABLE_kor
```

```python
df_53TABLE_col_list = df_53TABLE_col.to_dict('records')
new_dict_53 = {}

for df_1 in df_53TABLE_col_list:

    
    new_dict_53[df_1['변수명']] = df_1['표준한글컬럼명']        
new_dict_53
    

df_53TABLE_kor = df_53TABLE.rename(columns = new_dict_53)
df_53TABLE_kor
```

```python
df_20TABLE_kor.to_excel('./files_new/df_20Table_kor_명세서일반내역.xlsx', index=False)
df_30TABLE_kor.to_excel('./files_new/df_30Table_kor_진료내역.xlsx', index=False)
df_40TABLE_kor.to_excel('./files_new/df_40Table_kor_상병.xlsx', index=False)
df_53TABLE_kor.to_excel('./files_new/df_53Table_kor_처방전상세내역.xlsx', index=False)
```
