```python
tips = sns.load_dataset("tips")
print(tips.head())
print(type(tips))
 
```

```python
tips.info()
```

```python
fig=plt.figure()
axes1=fig.add_subplot(1,1,1)
fig=plt.figure(figsize=(20,20),dpi=600)
```

```python
axes1.hist(tips['total_bill'], bins=10)
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')
```

```python
scatter_plot = plt.figure()
axes1= scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'],tips['tip'])
axes1.set_title('Histogram of Total Bill vs Tip')
axes1.set_xlabel('Total Bill_')
axes1.set_ylabel('TIp')
```

```python
plt.scatter(tips['total_bill'],tips['tip'],c='red')
```

```python
boxplot= plt.figure()

axes1.boxplot(
    [tips[tips['sex']=='Female']['tip'],
    tips[tips['sex'] == 'Male']['tip']],
    labels=['Female','Male'])
axes1.set_xlabel('Gender')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Gender')
```

```python
def recode_gender(gender):
    if gender =='Female':
        return 'red'
    else:
        return 'blue'
```

```python
tips['color'] = tips['sex'].apply(recode_gender)
tips.head()
```

```python
scatter_plot = plt. ()
axes1 =scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size']*10,
    c = tips['color'],
    alpha=0.5)

axes1.set_title('Total Bill vs Tip Colored by Gender and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

```python
bar_plot = plt.figure()
axes1 = bar_plot.add_subplot(1,1,1)

axes1.bar(tips['sex'],tips['total_bill'])
```

```python
import seaborn as sns
tips = sns.load_dataset("tips")
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram with Density Plot')
```

```python
ax = plt.subplots()
ax=sns.distplot(tips['total_bill'], kde =False)
ax.set_title('Total Bill Histogram')
```

```python
ax = plt.subplots()
ax=sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram')
```

```python
ax = plt.subplots()
ax= sns.regplot(x='total_bill', y= 'tip', data =tips)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax = plt.subplots()
ax= sns.regplot(x='total_bill', y= 'tip', data =tips)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')

ax = sns.regplot(x = 'total_bill', y= 'tip',
data = tips, fit_reg=False)
```

```python
joint = sns.jointplot(x = 'total_bill', y ='tip', data = tips)
joint.set_axis_labels(xlabel = 'Total Bill',ylabel = 'Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize = 10,
y = 1.03)
```

```python
hexbin = sns.jointplot(x = 'total_bill', y ='tip', data = tips, kind = "hex")
hexbin.set_axis_labels(xlabel = 'Total Bill',ylabel = 'Tip')
hexbin.fig.suptitle('hexbin Plot of Total Bill and Tip', fontsize = 10,
y = 1.03)
```

```python
ax = plt.subplots()
ax = sns.kdeplot(x = tips['total_bill'], y=tips['tip'], shade = True)  # shade =  색상차이(명암유무)
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax = plt.subplots()
ax= sns.violinplot(x = 'time', y = 'total_bill', data = tips)
ax.set_title('Violinplot of total bill by time of day')
ax.set_xlabel('날짜')
ax.set_ylabel('Total Bill')
```

```python
fig = sns.pairplot(tips)  # seaborn을 쓰는 이유이기도함 가장 많이 씀
```

```python
pair_gird = sns.PairGrid(tips)
pair_gird=pair_grid.map_upper(sns.regplot)
pair_gird=pair_grid.map_lower(sns.kdeplot)
pair_grid=pair_grid.map_diag(sns.distplotm,rug = True)
plt.show()
```
