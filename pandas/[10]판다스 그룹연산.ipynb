```python
def my_mean(values):
    n = len(values)
    sum = 0
    for value in values:
        sum += value
    return sum / n
```

```python
df.groupby('year').lifeExp.agg(my_mean)
```

```python
def my_mean_diff(values, diff_value):
    n = len(values)
    sum = 0
    for value in values:
        sum += value
    mean = sum / n
    return mean - diff_value
```

```python
global_mean = df['lifeExp'].mean()
global_mean
```

```python
df.groupby('year').lifeExp.agg(
my_mean_diff, diff_value=global_mean)
```

```python
df.groupby('year').lifeExp.agg(
[np.count_nonzero, np.mean, np.std])
```

```python
df.groupby('year').agg(
    {'lifeExp':'mean', 'pop':'std', 'gdpPercap':'mean'})
```

```python
def my_zscore(x):
    return (x - x.mean()) / x.std()
```

```python
df['zscore'] = df.groupby('year').lifeExp.transform(my_zscore)
```

```python
import seaborn as sns
import numpy as np
```

```python
np.random.seed(1234)
tips_10 = sns.load_dataset('tips').sample(10)
tips_10
```

```python
tips_10.loc[np.random.permutation(tips_10.index)[:4], 'total_bill'] = np.NaN
```

```python
tips_10
```

```python
tips_10.loc[[31, 52, 71, 5], 'total_bill'] = np.NaN
tips_10
```

```python
def fill_na_mean(x):
    avg = x.mean()
    return x.fillna(avg)
```

```python
tips_10['fill_total_bill'] = tips_10.groupby(['sex', 'time']).total_bill.transform(fill_na_mean)
```
